### HTML (25문제)

#### **HTML 태그 기본 및 구성요소**
1.  HTML 요소(Element)를 구성하는 3가지 기본 요소를 설명하시오.
상위태그, 마지막 태그, 그 안에 하위태그

2.  HTML 태그 내에서 속성(Attribute)의 역할은 무엇이며, 어떤 형식으로 작성되는지 설명하시오.
속성은 태그의 부가 정보를 제공하며, 이름="값"형식으로 작성된다.

3.  모든 HTML 문서의 구조를 잡는 최상위 태그는 무엇이며, 문서의 기본 메타데이터를 정의하는 영역과 실제 콘텐츠를 표시하는 영역을 나누는 두 개의 핵심 태그를 설명하시오.
최상위 태그는 html이고, head가 메타데이터, body가 콘텐츠를 표시한다.

4.  내용(content)을 가지지 않는 빈 태그(Empty Tag)의 개념을 설명하고, 대표적인 예시 2가지를 드시오.
빈 태그는 내용이 없고, 닫는 태그가 없다.(/) 그리고 에시로는 문단을 나누는 br과 구분선을 추가하는 hr이 있다.

5.  `<title>` 태그의 기능과 이 태그가 HTML 문서의 어느 부분에 위치해야 하는지 설명하시오.
브라우저 탭의 제목이다. head안에 위치해야 한다.

#### **제목(h1~h6), 구분선(hr), 줄바꿈(br)**
6.  `<h1>` 태그의 의미론적(semantic) 중요성에 대해 설명하시오.
문서의 최상위 제목이고, 가장 중요한 제목(대표제목)을 나타날때 사용한다.

7.  `<h1>`부터 `<h6>` 태그까지의 관계는 무엇이며, 검색 엔진 최적화(SEO) 관점에서 이들의 올바른 사용법은 무엇인지 설명하시오.
폰트 사이즈에 따라 달라진다. 6으로 갈수록 사이즈가 작아진다. h1->h6 계층적 구조 순서대로 유지해야 검색엔진의 가독성이 올라간다.

8.  `<hr>` 태그의 시각적 효과와 의미론적 역할을 설명하시오.
html에 구분선을 추가한다. 주제나 문단을 구분할 때 사용

9.  `<p>` 태그 내에서 단순히 줄만 바꾸고 싶을 때 `<br>` 태그를 사용하는 이유를 설명하시오.
p는 문단 단위로 나뉘는 거여서 줄만 바꾸고 싶을 때에는 br태그를 사용하는 것이 좋다.

#### **특수문자 엔티티 및 pre 태그**
10. HTML 문서 내에서 `<`, `>`, `&` 와 같은 문자를 브라우저가 태그가 아닌 일반 문자로 인식하게 하려면 어떻게 해야 하는지, 그 방법(문자 엔티티)의 개념을 설명하시오.
<, >, &를 일반 문자로 인식하게 하려면 특수문자 엔티티를 사용해야 한다. &lt;, &gt;, &amp; 와 같은 문자를 대체로 사용하는 것이다.

11. 웹 페이지에서 여러 개의 공백(space)을 연속으로 입력해도 왜 한 칸으로만 표시되는지 설명하고, 이를 해결하기 위한 문자 엔티티를 제시하시오.
HTML에서 공백은 &nbsp; 엔티티를 사용하여 연속으로 공백 추가할 수 있다.

12. `<pre>` 태그가 일반적인 HTML 태그와 구별되는 가장 큰 특징은 무엇이며, 어떤 상황에서 유용하게 사용되는지 설명하시오.
pre태그로 감싸면 어떤 내용이 들어가든 문자 그대로 보여준다. 코드나 이모티콘 같은 레이아웃을 유지할 때 유용하다. 

#### **인라인(Inline) / 블록(Block) 요소**
13. 인라인(Inline) 요소의 2가지 주요 특징을 설명하시오.
콘텐츠 크기만큼만 공간을 차지하고, 옆에 다른 요소를 배치 가능하다.

14. 블록(Block) 요소의 2가지 주요 특징을 설명하시오.
화면 가로 전체를 차지하고, 자동으로 줄바꿈이 된다.

15. `<span>` 태그와 `<div>` 태그를 각각 인라인/블록 요소로 구분하고, 각각의 주된 용도를 비교하여 설명하시오.
span태그는 인라인이고, 부분적으로 스타일을 적용하는 반면, div태그는 블록이고, 레이아웃을 구분하는데 사용한다.

블록 요소 안에 인라인 요소를 포함하는 것이 표준에 맞는 방식이다. 

#### **목록(ol, ul, dl) 태그**

ol태그로 li태그들을 감싼다.

### Oracle SQL (25문제)

#### **SELECT 문**

모든 커럶을 조회할 때 SELECT * 이런식으로 사용한다. 

27. `SELECT` 절에서 조회할 특정 컬럼들을 나열하는 기본 문법을 설명하시오.

AS 키워드를 사용한다. (생량 가능) EX) SELECT SAL as SALARY

29. 특정 컬럼의 데이터에 중복된 값이 많을 때, 중복을 제거하고 고유한 값들만 조회하기 위해 사용하는 키워드는 무엇이며, 어떻게 사용하는지 설명하시오.

30. `SELECT` 절에서 산술 연산자(+, -, *, /)를 사용하여 기존 컬럼의 값을 가공하거나 계산한 결과를 새로운 컬럼처럼 조회하는 방법을 설명하시오.

=, >, <, >=, <= 조건식을 만들 때 사용한다. (같다, 크다, 작다, 크거나 같다, 작거나 크다 등등)

작은 따옴표를 사용해야한다.

WHERE 조건문 AND 조건문으로 사용

WHERE 조건문 OR 조건문으로 사용

36. 특정 컬럼의 값이 비어있는지(NULL) 여부를 확인할 때 `=` 연산자 대신 `IS NULL` 또는 `IS NOT NULL`을 사용해야 하는 이유를 설명하시오.
NULL은 값이 아니기 때문에 =연산자로는 표시할 수 없다.

37. `LIKE` 연산자는 어떤 경우에 사용되며, 모든 문자를 의미하는 와일드카드 `%`와 한 글자를 의미하는 `_`의 차이점을 설명하시오.
LIKE 연산자는 와일드카드를 사용하여 특정 값을 조회하는데 사용한다. %는 임의 길이의 문자열을 대체하는거고, _는 한 글자만 대체한다.

38. `ENAME`이 'J'로 시작하고 'n'으로 끝나는 직원을 찾는 `WHERE` 절 조건을 `LIKE`를 사용하여 작성하시오.

### **Phase 4: 복합 조건을 이용한 고급 데이터 필터링 및 정렬 (36~50번)**

#### **Oracle SQL: 고급 WHERE절, 함수, 복합 ORDER BY 활용**
36. `PLANETS` 테이블에서 인구가 10억 이상 100억 이하인 행성을 모두 조회하세요. (`BETWEEN` 연산자 사용)
SELECT *
FROM PLANETS
WHERE population BETWEEN 1000000000 AND 10000000000;

37. `SPECIES` 테이블에서 모성이 'Earth'(ID: 101) 또는 'Mars'(ID: 102)인 종족을 모두 조회하세요. (`IN` 연산자 사용)
SELECT *
FROM SPECIES
WHERE home_planet_id IN (101, 102);

38. 대기 타입(`atmosphere_type`)이 'Breathable' 이면서 **(AND)** 인구가 0 이상인 행성, **또는(OR)** 대기 타입이 `NULL`인 모든 행성을 조회하는 복합 `WHERE` 절을 작성하세요. (괄호 `()`를 사용하여 조건의 우선순위를 명확히 해야 합니다.)
SELECT *
FROM PLANETS
WHERE (atmosphere_type = 'Breathable' AND population >= 0)
   OR atmosphere_type IS NULL;

39. `SELECT` 문에서 `planet_name`과 `population`을 조회하되, `population`이 `NULL`일 경우 0으로 표시하여 조회하는 `NVL` 함수를 사용한 쿼리를 작성하세요.
SELECT planet_name, NVL(population, 0) AS population
FROM PLANETS;

40. `STAR_SYSTEMS` 테이블에서 `system_name`을 모두 **소문자**로 변환하여 조회하는 `LOWER` 함수를 사용한 쿼리를 작성하세요.
SELECT LOWER(system_name) AS system_name_lower
FROM STAR_SYSTEMS;

41. `PLANETS` 테이블에서 행성 이름(`planet_name`)과 그 이름의 글자 수를 함께 조회하는 `LENGTH` 함수를 사용한 쿼리를 작성하세요.
SELECT planet_name, LENGTH(planet_name) AS name_length
FROM PLANETS;

42. `PLANETS` 테이블에서 `discovery_date`를 기준으로 **최신순(내림차순)** 으로 정렬하되, 만약 발견일이 같다면 `planet_name`의 **알파벳순(오름차순)** 으로 2차 정렬하는 `ORDER BY` 절을 작성하세요.
SELECT *
FROM PLANETS
ORDER BY discovery_date DESC, planet_name ASC;

43. `SELECT` 조회 시, `planet_name` 과 `'(ID: ' || planet_id || ')'` 를 연결하여 "Earth (ID: 101)" 과 같은 형태로 보이도록 'FormattedName' 이라는 별칭을 가진 컬럼을 만드는 쿼리를 작성하세요.
SELECT planet_name || ' (ID: ' || planet_id || ')' AS FormattedName
FROM PLANETS;

44. `PLANETS` 테이블에서 소속 항성계(`system_id`)의 종류가 몇 개인지 **중복을 제거**하여 조회하는 쿼리를 작성하세요. (`DISTINCT` 사용)
SELECT DISTINCT system_id
FROM PLANETS;

45. `SPECIES` 테이블에서 종족 이름이 5글자 이하인 종족들만 조회하는 쿼리를 `LENGTH` 함수와 `WHERE` 절을 조합하여 작성하세요.
SELECT *
FROM SPECIES
WHERE LENGTH(species_name) <= 5;

46. `PLANETS` 테이블의 `population`을 1,000,000으로 나눈 값(백만 단위 인구)을 기준으로 내림차순 정렬하는 `ORDER BY` 절을 작성하세요.
SELECT *
FROM PLANETS
ORDER BY (population / 1000000) DESC;

47. `WHERE` 절에서 `LIKE` 와일드카드 `_`(언더스코어)를 사용하여, 이름이 정확히 4글자인 행성만 조회하는 쿼리를 작성하세요.

48. `UPDATE` 문에 `WHERE` 절을 사용하지 않았을 때 발생할 수 있는 치명적인 문제점은 무엇인지 설명하세요.
UPDATE문에 WHERE절을 사용하지 않으면, 모든 항목이 한꺼번에 갱신되어 데이터 전체의 손실 위험성이 있다.

49. `DELETE` 문에 `WHERE` 절을 사용하여, `discovery_date`가 2000년 1월 1일 이전인 오래된 행성 기록만 삭제하는 쿼리를 작성하세요. (날짜 형식: `TO_DATE('2000-01-01', 'YYYY-MM-DD')`)
DELETE FROM PLANETS
WHERE discovery_date < TO_DATE('2000-01-01', 'YYYY-MM-DD'); 


50. 최종 미션: `PLANETS` 테이블에서 `system_id`가 1이고 대기가 'Breathable'인 행성들의 이름과 인구를 조회하여, 인구가 많은 순으로 정렬하는 최종 보고용 쿼리를 작성하세요. 이 쿼리의 결과를 어떤 HTML 태그(테이블, 목록 등)로 표현하는 것이 가장 효과적일지 설명하며 프로젝트를 마무리하세요.
SELECT planet_name, population
FROM PLANETS
WHERE system_id = 1
  AND atmosphere_type = 'Breathable'
ORDER BY population DESC;

이 쿼리는 행성 이름과 인구의 열 구조이기 떄문에 table로 표현 하는 것이 가장 효과적일 것이다.

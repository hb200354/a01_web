8. [개념] 오라클에서 트랜잭션이란 무엇이며, 트랜잭션 처리를 위한 핵심명령어 3개와 그 기능에 대하여 기본예제로 설명하세요.
데이터베이스 작업을 처리하는 하나의 논리적 단위를 트랜잭션이라고 한다.
핵심명령어에는 COMMIT, ROLLBACK, SAVEPOINT가 있다.
커밋은 변경된 내용을 영구적으로 DB에 저장하는 명령어이고, ROLLBACK은 작업을 취소하고 싶을 때 가장 마지막에 커밋된
지정으로 돌아가는거, SAVEPOINT는 중간 저장 지점을 만들어서 롤백했을 때 해당 SAVEPOINT로 돌아갈 수 있다.
EX)
SELECT * FROM HERO_LIST;
UPDATE HERO_LIST
    SET POWER = 0
WHERE HERO_NAME = '아이언맨';
ROLLBACK;
-- SAVEPOINT : 임시 저장 지점 설정
SAVEPOINT IRONMAN_RETIRED;
DELETE FROM HERO_LIST;
--모든 데이터 삭제
SELECT * FROM HERO_LIST;
ROLLBACK TO IRONMAN_RETIRED;
COMMIT;

3. [개념] NULLIF의 기본 기능을 예제를 통해 기술하고, 실제 사용하는 용도를 설명하세요.
기본 기능: 데이터 1과 데이터 2가 같으면 NULL, 다르면 데이터 1로 처리
SELECT ENAME, SAL, COMM, NULLIF(COMM, 0) COMM_RESULT
FROM EMP; --> COMM값이 0과 같으면 NULL인것이므로, 0을 NULL로 바꿔줌
실제 사용 용도: 위의 예제처럼 특정값을 NULL로 변환할 때나, 계산할 때 0으로 나누는 오류를 방지할 때 사용한다.

 나타낼려고 한다. 오류가 나오지 않게 NULLIF를 사용하세요

6. [실습]
   로그인 폼의 두 가지 버튼 구현하기
   문제 설명:
   로그인 폼에는 아이디와 비밀번호를 서버로 '전송'하는 버튼과, 전송 없이 화면에서 '도움말'을 보여주는 버튼이 필요합니다. <input> 태그와 <button> 태그를 각각 사용하여 이 두 가지 기능의 버튼을 모두 구현해 보세요.
   지시사항:
   <form> 태그 내부에 아이디와 비밀번호를 입력받는 <input type="text">, <input type="password">를 만듭니다.
   "로그인" 버튼:
   클릭 시 폼의 내용이 **전송(submit)**되어야 합니다.
   <input> 태그를 사용하여 만듭니다.
   "도움말 보기" 버튼:
   클릭 시 폼이 전송되지 않고, alert('아이디와 비밀번호를 입력해주세요.'); 라는 경고창만 떠야 합니다.
   <button> 태그를 사용하여 만듭니다.

<!DOCTYPE html>
<html lang="ko">
<head>
    <meta charset="UTF-8">
    <meta name="viewport" content="width=device-width, initial-scale=1.0">
    <title>로그인 폼</title>
    <script>
        function showHelp() {
            alert('아이디와 비밀번호를 입력해주세요.');
        }
    </script>
</head>
<body>
    <h3>로그인 폼</h3>
    <hr>
    <form>
        <label for="userid">아이디:</label>
        <input type="text" id="userid" name="userid" placeholder="아이디 입력"><br><br>

        <label for="pwd">비밀번호:</label>
        <input type="password" id="pwd" name="pwd" placeholder="비밀번호 입력"><br><br>

        <!-- 로그인 버튼 (폼 전송) -->
        <input type="submit" value="로그인">

        <!-- 도움말 버튼 (폼 전송 없이 alert) -->
        <button type="button" onclick="showHelp()">도움말 보기</button>
    </form>
</body>
</html>

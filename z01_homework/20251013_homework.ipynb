1. [개념] select의 기본 코드와 속성을 기본예제와 함께 설명하세요
<select> 태그로 감싸고, <option> 태그로 선택 항목을 정의한다.
속성: 
   - <select>: 콤보박스 전체를 감싸는 태그
   - <option>: 콤보박스 내의 각각의 선택 항목을 정의하는 태그
   - <optgroup>: 관련된 옵션들을 그룹화할 때 사용하는 태그
   - name 속성: 폼 데이터 전송 시 선택된 값을 식별하는 데 사용
   - value 속성: 각 옵션의 실제 값을 지정
   - selected 속성: 기본적으로 선택된 옵션을 지정
   - multiple 속성: 사용자가 여러 항목을 동시에 선택할 수 있도록 허용
   - label 태그: optgroup에 레이블을 제공하여 그룹의 의미를 명확히 함
예제)
<form action="result.jsp" method="post">
  <label for="fruit">좋아하는 과일:</label>
  <select name="fruit" id="fruit">
    <option value="apple">사과</option>
    <option value="banana">바나나</option>
    <option value="grape">포도</option>
  </select>
  <input type="submit" value="제출">
</form>

4. [개념] RTRIM, LTRIM, TRIM의 차이점을 기본 예제를 통해서 설명하세요.
LTRIM(Left TRIM) : 왼쪽의 불필요한 부분 잘라냄
RTRIM(Right TRIM) : 오른쪽의 불필요한 부분 잘라냄
TRIM : 양쪽 모두를 잘라내거나, 특정 방향의 특정 문자 하나를 지정해서 잘라냄

SELECT LTRIM('***HELLO', '*') FROM DUAL; -- 결과: HELLO
SELECT RTRIM('HELLO***', '*') FROM DUAL; -- 결과: HELLO
SELECT TRIM('=' FROM '=HELLO=') FROM DUAL; -- 결과: HELLO

### **데이터베이스 (Database)**

1.  **[개념] 날짜 데이터 연산**
    날짜와 숫자를 더하거나 빼는 연산 방식을 예제를 통해 설명하세요. (참고: `SYSDATE + 0.5`는 현재 시간으로부터 12시간 후를 의미합니다.)

날짜형 데이터에 숫자를 더하거나 빼면, 일 단위(day) 로 계산됩니다. 
ex) +1 = 내일, -1 = 어제, +0.5 = 12시간 후
SELECT SYSDATE AS 현재날짜,
       SYSDATE + 1 AS 내일,
       SYSDATE - 1 AS 어제,
       SYSDATE + 0.5 AS 12시간후
FROM DUAL;


2.  **[개념] 숫자 처리 함수와 날짜 계산**
    `FLOOR()`, `CEIL()`, `ROUND()` 함수의 의미를 날짜 계산 예제를 통해 설명하세요.

FLOOR(): 소수점 아래 버림
CEIL: 소수점 올림
ROUND: 반올림
EX)
SELECT ENAME, HIREDATE,
       FLOOR(SYSDATE - HIREDATE) AS 버림,
       CEIL(SYSDATE - HIREDATE) AS 올림,
       ROUND(SYSDATE - HIREDATE) AS 반올림
FROM EMP;


3.  **[실습] D-Day 계산**
    오늘을 기준으로 수료일(2026-03-25)까지 며칠이 남았는지 계산하세요.

SELECT TO_DATE('2026-03-25', 'YYYY-MM-DD') - SYSDATE AS 남은일수
FROM DUAL;

4.  **[실습] 근무일수 계산**
    사원 테이블(EMP)을 활용하여 오늘(SYSDATE)을 기준으로 각 사원의 근무일수를 계산하고, ‘사원명’, ‘입사일’, ‘근무일수’ 컬럼으로 출력하세요.

SELECT ENAME AS 사원명,
       HIREDATE AS 입사일,
       FLOOR(SYSDATE - HIREDATE) AS 근무일수
FROM EMP;

5.  **[실습] 특정 월 입사자 조회**
    사원 테이블(EMP)을 활용하여 입사일이 1월에서 5월 사이인 사원을 검색하여 출력하세요.

SELECT ENAME, HIREDATE
FROM EMP
WHERE TO_NUMBER(TO_CHAR(HIREDATE, 'MM')) BETWEEN 1 AND 5;

6.  **[실습] 특정 연도 또는 월 입사자 조회**
    입사일에서 연도(YYYY)와 월(MM) 정보를 추출하여, 1987년에 입사했거나 또는 5월에 입사한 사원을 검색하여 출력하세요.

SELECT ENAME, HIREDATE
FROM EMP
WHERE TO_CHAR(HIREDATE, 'YYYY') = '1987'
   OR TO_CHAR(HIREDATE, 'MM') = '05';

7.  **[개념] 그룹 함수**
    그룹 함수를 사용하는 이유, 종류, 그리고 기본 형식을 예제를 통해 설명하세요.

SELECT 부서번호, AVG(SAL) AS 평균급여
FROM EMP
GROUP BY 부서번호;

8.  **[실습] 직책별 급여 분석**
    사원 테이블(EMP)에서 직책별 최고 급여와 평균 급여를 계산하여 출력하세요.

SELECT JOB AS 직책,
       MAX(SAL) AS 최고급여,
       ROUND(AVG(SAL), 2) AS 평균급여
FROM EMP
GROUP BY JOB;

9.  **[실습] 조건별 그룹 급여 분석**
    사원번호(EMPNO)가 홀수인 그룹과 짝수인 그룹으로 나누어, 각 그룹의 최고 급여와 최저 급여를 `MOD()` 함수를 사용하여 출력하세요.

SELECT MOD(EMPNO, 2) AS 그룹,
       MAX(SAL) AS 최고급여,
       MIN(SAL) AS 최저급여
FROM EMP
GROUP BY MOD(EMPNO, 2);

10. **[개념] 그룹 함수의 범위**
    `GROUP BY` 절 없이 그룹 함수를 사용하는 경우와 `GROUP BY` 절과 함께 사용하는 경우의 차이점을 설명하세요.

GROUP BY 없이 그룹 함수 사용하는 경우는 전체 행을 하나의 그룹으로 집계할 때이고,
GROUP BY와 함께 사용하는 경우는 어떤 특정한 기준의 컬럼별로 그룹화 할때이다.

11. **[실습] NULL 값을 포함한 평균 계산**
    사원 테이블(EMP)에서 전체 사원의 평균 보너스(COMM)를 구하려고 합니다. 보너스를 받지 않는 사원(COMM이 NULL)을 포함하여 정확한 평균을 계산하려면 어떻게 처리해야 할까요?

* NVL 함수를 사용하면 된다.
SELECT AVG(NVL(COMM, 0)) AS 평균보너스
FROM EMP;

1.  **[개념] 날짜 및 시간 관련 입력 타입**

종류와 속성: 
    type="month" : 특정 월을 선택할 때 사용

<input type="date" name="birthday">
<input type="time" name="alarm">
<input type="datetime-local" name="meeting">

9.  **[개념] 파일 업로드 폼**
    HTML 폼을 사용하여 파일을 업로드할 때 필요한 기본 코드 구성과 주요 속성(`enctype`, `accept`, `multiple` 등)을 예제를 통해 설명하세요.

주요 속성:
    enctype="mutipart/form-data" : 파일 업로드 인코딩
    accept : 허용 파일 형식 지정
    multiple : 여러 파일 선택 가능
ex)
<form action="upload.php" enctype="multipart/form-data">
  <input type="file" name="file" accept="image/*" multiple="multiple">
  <button type="submit">업로드</button>
</form>

    여러 이미지 파일을 동시에 업로드할 수 있는 전체 폼을 완성하세요.
    * 파일 전송을 위해 `<form>` 태그에 필수 속성(`enctype="multipart/form-data"`)을 추가하세요.
    * '갤러리 이미지 선택'이라는 텍스트를 가진 `<label>`을 파일 선택 버튼과 연결하세요.
    * 파일 선택 버튼(`<input type="file">`)은 모든 종류의 이미지(`image/*`)를 여러 장 동시에 선택할 수 있도록 설정하세요.
    * 마지막으로 '업로드'라는 텍스트가 표시된 제출 버튼을 만드세요.
  
<form action="upload.php" method="post" enctype="multipart/form-data">
  <label for="gallery">갤러리 이미지 선택</label><br>
  <input type="file" id="gallery" name="images" accept="image/*" multiple><br><br>
  <button type="submit">업로드</button>
</form>

   특히, TRUNC, FLOOR의 차이점을 매개변수와 함께 자세히 설명하세요

   CEIL:  수를 올림함
   FLOOR: 수를 내림함

2. [실습] 아래내용을 위 수학적 함수를 이용하여 SQL로 처리하세요
   1) 서울의 아침 최저 기온은 영하 8도였고, 같은 시각 부산의 기온은 영상 5도였습니다. 두 도시의 기온 차이는 절대값으로 몇 도입니까?(ABS)
    
    SELECT ABS(-8 - 5) "기온차"
    FROM DUAL;
    * 13℃

   2) 연필 180자루를 25명의 학생에게 똑같이 나누어 주려고 합니다. 한 학생당 최대 몇 자루의 연필을 받을 수 있으며, 나누어 주고 남는 연필은 총 몇 자루일까요?(TRUNC, MOD)
    
    SELECT TRUNC(180 / 25) "한명당연필수",
       MOD(180, 25) "남은연필"
    FROM DUAL;  
    * 한 명당 최대 7자루, 남는 연필 5자루

   3) 어떤 프로젝트를 완성하는 데 총 300일이 걸립니다. 이를 주 단위와 일 단위로 변환하면 총 몇 주 하고 며칠이 걸리는 것일까요? (단, 1주는 7일로 계산합니다.(TRUNC, MOD)
    
    SELECT TRUNC(300 / 7) "총주수",
       MOD(300, 7) "남은일수"
    FROM DUAL;
    * 총 42주 6일

   4) 회사 워크숍을 가기 위해 10인승 차량을 예약해야 합니다. 20번 부서의 직원은 총 5명입니다. `5 / 10 = 0.5` 이지만, 0.5대의 차량은 없으므로 무조건 올려서 1대의 차량을 예약해야 합니다.(CEIL)
    
    SELECT CEIL(5 / 10) "필요차량대수"
    FROM DUAL;
 
   5) 경리부에서 모든 직원의 월급(SAL)에 대해 3.3%의 소득세를 원천징수해야 합니다. 세금 계산 시 소수점이 발생하면 정수 단위로 반올림하여 최종 세액을 결정합니다.(ROUND)
   속성 처리: 월급 * 0.033으로 계산된 세금에서 소수점 첫째 자리를 반올림.  
    
    SELECT SAL,
       ROUND(SAL * 0.033, 0) AS "세금"
    FROM EMP;

6. [개념] MONTHS_BETWEEN() 함수는 매개변수는 어떤 것이 있고, 결과는 어떤 처리를 하는지 기술하세요

    MONTHS_BETWEEN() 함수는 두 날짜 사이의 개월 수 차이를 계산한다.
    매개변수로는 두 개의 날짜(date1, date2) 를 받으며,
    date1 - date2 의 개월 수 결과를 반환한다.

7. [실습] 각 직원의 특정한 기념일1(1990-01-01), 기념일2(2000-12-01)에 개월수 계산하여 출력하세요 [TO_DATE(), MONTHS_BETWEEN 활용]

    SELECT TO_DATE('1990-01-01','YYYY-MM-DD') AS 기념일1,
       TO_DATE('2000-12-01','YYYY-MM-DD') AS 기념일2,
       MONTHS_BETWEEN(
         TO_DATE('2000-12-01','YYYY-MM-DD'),
         TO_DATE('1990-01-01','YYYY-MM-DD')
       ) AS "개월수차이"
    FROM DUAL;

INNER JOIN - SELECT E.ENAME, D.DNAME FROM EMP E JOIN DEPT D ON E.DEPTNO = D.DEPTNO;
LEFT OUTER JOIN - SELECT E.ENAME, D.DNAME FROM EMP E LEFT OUTER JOIN DEPT D ON E.DEPTNO = D.DEPTNO;
RIGHT OUTER JOIN - SELECT E.ENAME, D.DNAME FROM EMP E RIGHT OUTER JOIN DEPT D ON E.DEPTNO = D.DEPTNO;

4. [개념] left outer join, right outer join의 궁극적인 차이점을 예제와 함께 설명하세요

LEFT OUTER JOIN -> 왼쪽 테이블의 모든 행 포함
RIGHT OUTER JOIN -> 오른쪽 테이블의 모든 행 포함

EX)
-- LEFT OUTER JOIN
SELECT i.Name, g.GroupName
FROM Idols i LEFT JOIN Groups g ON i.GroupID = g.GroupID;

-- RIGHT OUTER JOIN
SELECT i.Name, g.GroupName
FROM Idols i RIGHT JOIN Groups g ON i.GroupID = g.GroupID;

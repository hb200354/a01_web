13.  **(월별 입사자 통계)** 매월 입사한 사원 수를 분석하려고 합니다. `EMP` 테이블에서 월별 입사자 수를 조회하는 쿼리를 작성하세요.

SELECT TO_CHAR(HIREDATE,'YYYY-MM') AS MONTH, COUNT(*)

14.  **(급여 등급 분류)** `EMP` 테이블에서 급여(`SAL`)가 3000 이상이면 'A등급', 2000 이상이면 'B등급', 1000 이상이면 'C등급', 그 외는 'D등급'으로 분류하여 사원명과 함께 '급여등급' 컬럼으로 출력하세요.

SELECT ENAME,
  CASE 
    WHEN SAL >= 3000 THEN 'A등급'
    WHEN SAL >= 2000 THEN 'B등급'
    WHEN SAL >= 1000 THEN 'C등급'
    ELSE 'D등급'
  END AS "급여등급"
FROM EMP;

FROM EMP
GROUP BY JOB

1. [개념] foreign key 가 무엇인지? 생성방법을 기본 예제를 통하여 실습을 해보고 설명하세요.

foreign key는 두 테이블을 관계로 묶는 제약조건이다.
ex)
CREATE TABLE parent_table (
  id NUMBER PRIMARY KEY,
  name VARCHAR2(20)
);

CREATE TABLE child_table (
  child_id NUMBER PRIMARY KEY,
  parent_id NUMBER,
  CONSTRAINT fk_parent FOREIGN KEY (parent_id)
  REFERENCES parent_table(id)
);

2. [실습] 
   `PRODUCTS21` (제품) 테이블과 `ORDERS21` (주문) 테이블을 만드세요.
  `PRODUCTS21` 테이블: `product_code` (문자형, 5자리, PK)와 `product_name` (문자형) 컬럼을 가집니다.
  `ORDERS21` 테이블: `order_id` (숫자형, PK), `order_product` (문자형, 5자리), `quantity` (숫자형) 컬럼을 가집니다.
  `ORDERS21` 테이블의 `order_product` 컬럼이 `PRODUCTS` 테이블의 `product_code`를 참조하는 외래 키(FK)를 설정하세요.

CREATE TABLE PRODUCTS21 (
  product_code CHAR(5) PRIMARY KEY,
  product_name VARCHAR2(50)
);

CREATE TABLE ORDERS21 (
  order_id NUMBER PRIMARY KEY,
  order_product CHAR(5),
  quantity NUMBER,
  CONSTRAINT fk_order_product FOREIGN KEY (order_product)
  REFERENCES PRODUCTS21(product_code)
);

  'STUDENT' 사용자에게 부여했던 `CREATE TABLE` 권한(또는 `RESOURCE` 롤)을 다시 철회(Revoke)하세요.

-- 테이블 생성 권한 부여 (RESOURCE 롤 또는 개별 권한)

-- 권한 철회

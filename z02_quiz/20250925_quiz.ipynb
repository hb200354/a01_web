### 객관식 문제 (Multiple Choice)

**1. HTML 인라인(inline) 요소의 특징으로 옳지 않은 것은?**<br>
a) 다른 요소와 한 줄에 나란히 배치된다.<br>
b) 너비(width)와 높이(height) 속성을 직접 지정할 수 없다.<br>
<u>c) 내용물의 크기와 상관없이 항상 가로 한 줄을 전부 차지한다.</u><br>
d) `<span>`, `<a>`, `<strong>` 태그가 대표적인 인라인 요소이다.

**2. HTML 블럭(block) 요소에 대한 설명으로 가장 적절한 것은?**<br>

b) 상하 여백(margin-top, margin-bottom)을 적용할 수 없다.<br>
c) 다른 블럭 태그나 인라인 태그를 포함할 수 없다.<br>
<u>d) 항상 새로운 줄에서 시작하며, 해당 줄의 모든 너비를 차지한다.</u>

**3. 다음 중 블럭(block) 태그에 해당하지 않는 것은?**<br>

### 객관식 (4지 선다형, 20문제)

**HTML 테이블 관련**

1.  HTML 테이블에서 가로 방향(열)으로 여러 칸을 합칠 때 사용하는 속성은 무엇인가요?<br>
    ① `border`<br>
    ② `rowspan`<br>
    <u>③ `colspan`</u><br>
    ④ `align`

2.  `<td rowspan="3">` 코드가 의미하는 바로 가장 올바른 것은?<br>
    ① 현재 칸을 포함하여 오른쪽으로 총 3칸을 합친다.<br>
    <u>② 현재 칸을 포함하여 아래쪽으로 총 3칸(행)을 합친다.</u><br>
    ③ 테이블의 테두리 두께를 3으로 지정한다.<br>
    ④ 테이블의 높이를 300px로 지정한다.

3.  다음 중 테이블의 구조 제어(셀 병합)와 가장 관련이 깊은 속성은 무엇인가요?<br>
    ① `bgcolor`<br>
    ② `width`<br>
    <u>③ `colspan`</u><br>
    ④ `border`

4.  테이블의 배경색을 지정할 때 사용하는 속성은 무엇인가요?<br>
    ① `align`<br>
    <u>② `bgcolor`</u><br>
    ③ `color`<br>
    ④ `background`

5.  다음 중 과거에 테이블 디자인을 위해 사용되었으나, 현재는 CSS 사용이 강력히 권장되는 속성이 **아닌** 것은?<br>
    ① `border`<br>
    ② `width`<br>
    <u>③ `rowspan`</u><br>
    ④ `align`

6.  테이블의 너비와 높이를 지정하는 속성으로 올바르게 짝지어진 것은?<br>
    ① `size`, `length`<br>
    <u>② `width`, `height`</u><br>
    ③ `colspan`, `rowspan`<br>
    ④ `pixel`, `point`

7.  `<table align="center">` 속성은 테이블에 어떤 효과를 주나요?<br>
    ① 테이블 안의 텍스트를 가운데 정렬한다.<br>
    <u>② 테이블 자체를 웹 페이지의 중앙에 위치시킨다.</u><br>
    ③ 테이블의 모든 셀을 병합한다.<br>
    ④ 테이블의 테두리를 중앙선 스타일로 바꾼다.

8.  아래 HTML 코드에서 첫 번째 행(tr)은 총 몇 개의 열(column) 너비를 차지할까요?<br>

    ```html
    <table border="1">
        <tr><td colspan="3">A</td></tr>
        <tr><td>B</td><td>C</td><td>D</td></tr>
    </table>
    ```

    ① 1개<br>
    ② 2개<br>
    <u>③ 3개</u><br>
    ④ 4개

9.  웹 표준에 따라 테이블의 디자인(테두리, 색상, 정렬 등)을 제어할 때 가장 권장되는 방법은 무엇인가요?<br>
    ① HTML 속성을 최대한 많이 사용한다.<br>
    ② JavaScript를 사용한다.<br>
    <u>③ CSS(Cascading Style Sheets)를 사용한다.</u><br>
    ④ 테이블 내부에 주석으로 스타일을 정의한다.

10. `<td colspan="2">` 와 `<td rowspan="2">` 에 대한 설명으로 올바른 것은?<br>
    ① `colspan`은 세로, `rowspan`은 가로로 칸을 합친다.<br>

**Oracle SQL 관련**

11. `INSERT` 문을 사용하여 테이블의 **모든 컬럼**에 순서대로 데이터를 입력하는 올바른 형식은 무엇인가요?<br>
    ① `INSERT INTO 테이블명 (데이터1, 데이터2);`<br>
    ② `INSERT VALUES 테이블명 (데이터1, 데이터2);`<br>
    <u>③ `INSERT INTO 테이블명 VALUES (데이터1, 데이터2);`</u><br>
    ④ `INSERT 테이블명 (데이터1, 데이터2);`

12. `INSERT` 문에서 특정 컬럼만 지정하여 데이터를 입력할 때, 지정되지 않은 컬럼의 값은 어떻게 처리되나요?<br>
    ① 0으로 자동 입력된다.<br>
    ② 에러가 발생하여 입력이 취소된다.<br>
    ③ 공백(space)이 입력된다.<br>
    <u>④ `NULL` 값으로 처리된다.</u>

13. `EMP` 테이블의 구조와 데이터를 그대로 복사하여 `EMP_COPY`라는 새 테이블을 만드는 가장 빠른 방법은 무엇인가요?

    ② `COPY TABLE EMP TO EMP_COPY;`

    ④ `INSERT INTO EMP_COPY SELECT * FROM EMP;`

    ② `CREATE TABLE DEPT02 AS SELECT STRUCTURE FROM DEPT;`<br>
    ③ `CREATE TABLE DEPT02 FROM DEPT;`<br>
    ④ `CREATE TABLE DEPT02 AS SELECT * FROM DEPT;`

15. `DATE` 타입의 컬럼에 현재 데이터베이스 서버의 날짜와 시간을 입력할 때 사용하는 키워드는 무엇인가요?<br>
    ① `NOW()`<br>
    ② `TODAY`<br>
    ③ `CURRENT_TIME`<br>
    <u>④ `SYSDATE`</u>

16. `MEMBER` 테이블의 `id`와 `name` 컬럼에만 값을 입력하는 올바른 `INSERT` 구문은?<br>
    ① `INSERT INTO MEMBER VALUES ('userId', 'userName');`<br>

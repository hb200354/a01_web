24. **데이터베이스 트랜잭션 처리에서 `COMMIT`과 `ROLLBACK`의 역할을 비교하여 설명하시오.**
    COMMIT : 데이터를 확정하는 것. COMMIT 이후에는 원복이 되지 않는다.
    ROLLBACK : DML(INSERT, UPDATE, DELETE) 처리된 내용을 원복시키는 것.

25. **`SAVEPOINT`가 유용한 상황의 예를 한 가지 제시하고, `ROLLBACK TO [세이브포인트명]`을 사용하여 어떻게 문제를 해결할 수 있는지 설명하시오.**
    은행계좌 이체 시 
    1. A계좌에서 출금 ==> 성공한 직후, SAVEPOINT A_WITHDRAW_SUCCESS
    2. 거래 기록 남기기 ==> 만약 처리 도중 오류 발생 시 ROLLBACK TO A_WITHDRAW_SUCCESS 통해 복구 가능
    3. B계좌에 입금
    ==> 위와 같이 A계좌 출금 작업은 유지된 채로 오류가 난 작업만 다시 시도 가능해서 효율적.

11. DML 작업 후 변경된 내용을 데이터베이스에 영구적으로 저장하는 명령어는 무엇인가요?
    a) `ROLLBACK` : commit 전에는 수행되었는 insert, update, delete 등 내용을 수행 취소 가능
    b) `SAVEPOINT` : commit 전에 dml 내용 수행 시점을 지정.
                      insert 후, savepoint 지정1 설정
                      update 후, savepoint 지정2 설정

                      rollback to savepoing 지점2로 하면.. 
    c) `ALTER`
    #d) `COMMIT`

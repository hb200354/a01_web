13. 다음 중 `NVL` 함수의 특징으로 올바르지 **않은** 것은?
    ① 첫 번째 인자가 `NULL`일 경우 두 번째 인자를 반환한다.
    ② 첫 번째 인자와 두 번째 인자의 데이터 타입은 일치해야 한다.  
    	숫자형인 경우, null일 때 숫자형을 지정해주고, 문자열인 경우,  null일 경우 문자열을 지정해준다.
    	nvl(empno, 0)   : empno가 number(4,0) 이기에 0으로
    	nvl(ename, ' ') : ename이 varchar2 형이기에 ''문자열로
    	nvl(hiredate, sysdate) : hiredate가 date형이기에 sysdate로..
    #③ `SELECT` 문이나 `WHERE` 절에서 사용할 수 없다.
        select문에서도 주로 사용되고,
        where nvl(comm,0) = 0; 형식으로도 사용된다.
    ④ `NULL` 값이 포함된 숫자 연산에서 오류 없이 결과를 얻기 위해 사용된다.

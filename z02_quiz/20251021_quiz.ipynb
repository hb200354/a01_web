**3. 'SMITH'와 직무(JOB) 및 부서(DEPTNO)가 모두 동일한 다른 사원을 찾을 때, 메인 쿼리의 `WHERE` 절 조건으로 가장 적절한 것은 무엇인가요? (단, SMITH 제외)**<br>

③ `WHERE JOB IN (SELECT JOB FROM EMP WHERE ENAME = 'SMITH') OR DEPTNO IN (SELECT DEPTNO FROM EMP WHERE ENAME = 'SMITH')`<br>
④ `WHERE (JOB, DEPTNO) IN (SELECT JOB, DEPTNO FROM EMP) AND ENAME = 'SMITH'`

**4. 부서별(DEPTNO)로 급여(SAL)가 가장 많은 사원들의 정보를 출력하기 위한 쿼리의 `WHERE` 절로 올바른 것은 무엇인가요?**<br>

② `WHERE SAL IN (SELECT MAX(SAL) FROM EMP)`<br>
<u>③ `WHERE (DEPTNO, SAL) IN (SELECT DEPTNO, MAX(SAL) FROM EMP GROUP BY DEPTNO)`<br>
④ `WHERE (DEPTNO, SAL) = (SELECT DEPTNO, MAX(SAL) FROM EMP GROUP BY DEPTNO)`

**5. 오라클의 인라인 뷰(inline view)에 대한 설명으로 가장 적절한 것은 무엇인가요?**<br>

**6. 학습 내용에서 '프인 가별필'이라고 요약한 인라인 뷰의 특징 중, '별'이 의미하는 것은 무엇인가요?**<br>

**8. `EMP` 테이블에서 급여(SAL)를 가장 많이 받는 상위 3명을 올바르게 출력하는 쿼리는 무엇인가요?**<br>

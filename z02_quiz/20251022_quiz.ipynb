## 🖥️ HTML/CSS 및 Oracle SQL 연습 문제

### Ⅰ. 객관식 4지 선다형 (20문제)

**[HTML/CSS: 가상 클래스]**

1.  CSS 가상 클래스 중, 요소에 마우스 커서를 **올렸을 때** 스타일을 적용하는 선택자는 무엇인가?<br>
    ① `:active`<br>
    ② `:focus`<br>
    <u>③ `:hover`</u><br>
    ④ `:link`

2.  CSS 가상 클래스 중, 요소를 마우스로 **클릭하고 있는 동안** 스타일을 적용하는 선택자는 무엇인가?<br>
    <u>① `:active`</u><br>
    ② `:focus`<br>
    ③ `:hover`<br>
    ④ `:visited`

3.  주로 `<input>`이나 `<a>` 요소가 키보드 탭 등으로 선택(포커스)되었을 때 스타일을 적용하는 가상 클래스는 무엇인가?<br>
    ① `:active`<br>
    <u>② `:focus`</u><br>
    ③ `:link`<br>
    ④ `:first-child`

4.  `<a>` 태그의 링크 스타일 적용 시, 덮어쓰기 문제로 권장되는 순서(LVHA)로 올바른 것은?<br>
    <u>① `:link` → `:visited` → `:hover` → `:active`</u><br>
    ② `:visited` → `:link` → `:hover` → `:active`<br>
    ③ `:hover` → `:active` → `:link` → `:visited`<br>
    ④ `:link` → `:hover` → `:active` → `:visited`

5.  `<a>` 태그 가상 클래스 중, 사용자가 **한 번이라도 방문했던** 링크를 선택하는 것은 무엇인가?<br>
    ① `:link`<br>
    ② `:active`<br>
    ③ `:hover`<br>
    <u>④ `:visited`</u>

6.  구조적 가상 클래스 중, **형제 요소 중 첫째** 요소를 선택하는 것은 무엇인가?<br>
    ① `:nth-child(1)`<br>
    ② `:first-child`<br>
    ③ `:last-child`<br>
    <u>④ (①, ② 둘 다 해당)</u>

7.  형제 요소 중 짝수 번째 요소(2번째, 4번째...)를 모두 선택하는 선택자는 무엇인가?<br>
    ① `:nth-child(odd)`<br>
    <u>② `:nth-child(even)`</u><br>
    ③ `:nth-child(2n+1)`<br>
    ④ `:nth-child(2)`

8.  `<li>` 요소 중 마지막 요소를 제외한 모든 요소를 선택하는 방법으로 올바른 것은?<br>
    <u>① `li:not(:last-child)`</u><br>
    ② `li:first-child`<br>
    ③ `li:nth-child(-n+1)`<br>
    ④ `li:except(:last-child)`

9.  다음 중 `li:nth-child(3)` 선택자에 대한 설명으로 올바른 것은?<br>
    ① `li` 요소 중 3의 배수 번째 요소를 모두 선택한다.<br>
    ② `li` 요소 중 홀수 번째 요소를 모두 선택한다.<br>
    <u>③ `li` 요소 중 3번째 형제 요소를 선택한다.</u><br>
    ④ `li` 요소 중 3번째 자식 요소를 선택한다. (타입 무관)

10. 다음 중 구조적 위치 가상 클래스가 **아닌** 것은?<br>
    ① `:first-child`<br>
    <u>② `:focus`</u><br>
    ③ `:last-child`<br>
    ④ `:nth-child(odd)`

**[Oracle: 시퀀스 & 집합 연산자]**

11. 오라클에서 은행의 '번호표 발행기'처럼 자동으로 순서가 있는 번호를 생성해주는 객체는 무엇인가?<br>
    ① 테이블 (Table)<br>
    ② 뷰 (View)<br>
    ③ 인덱스 (Index)<br>
    ④ 시퀀스 (Sequence)

12. 시퀀스(Sequence)에서 **다음 번호**를 가져오면서 시퀀스 값을 **증가**시키는 명령어는 무엇인가?<br>
    ① `NEXTVAL`<br>
    ② `CURRVAL`<br>
    ③ `GETVAL`<br>
    ④ `INCVAL`

13. 시퀀스(Sequence)에서 현재 세션에서 마지막으로 **호출된 번호**를 (증가 없이) 다시 확인하는 명령어는 무엇인가?<br>
    ① `NEXTVAL`<br>
    ② `CURRVAL`<br>
    ③ `LASTVAL`<br>
    ④ `NOWVAL`

14. `CREATE SEQUENCE SEQ01 START WITH 100 INCREMENT BY 5;`로 시퀀스를 생성했다. `SEQ01.NEXTVAL`을 처음 호출했을 때 반환되는 값은?<br>
    ① 5<br>
    <u>② 100</u><br>
    ③ 101<br>
    ④ 105

15. 시퀀스 옵션 중, `MAXVALUE`에 도달했을 때 다시 `MINVALUE`부터 번호를 시작하도록 하는 옵션은 무엇인가?<br>
    ① `RESTART`<br>
    ② `NOCACHE`<br>
    <u>③ `CYCLE`</u><br>
    ④ `LOOP`

16. 두 개 이상의 SELECT 문 결과를 합칠 때, **중복된 행을 제거**하고 합집합을 반환하는 집합 연산자는 무엇인가?<br>
    ① `UNION`<br>
    ② `UNION ALL`<br>
    ③ `INTERSECT`<br>
    ④ `MINUS`

17. 두 개 이상의 SELECT 문 결과를 합칠 때, **중복된 행을 포함**하여 모든 결과를 반환하는 집합 연산자는 무엇인가?
    ① `UNION`
    ② `UNION ALL`
    ③ `INTERSECT`
    ④ `MINUS`

18. 두 SELECT 문 결과의 **공통된 행**(교집합)만을 반환하는 집합 연산자는 무엇인가?<br>
    ① `UNION`<br>
    ② `UNION ALL`<br>
    ③ `INTERSECT`<br>
    ④ `MINUS`

19. 첫 번째 SELECT 문 결과에서 두 번째 SELECT 문 결과에 **존재하는 행을 제외**한(차집합) 결과만을 반환하는 집합 연산자는 무엇인가?<br>
    ① `UNION`<br>
    ② `UNION ALL`<br>
    ③ `INTERSECT`<br>
    <u>④ `MINUS`</u>

20. 오라클 집합 연산자를 사용하기 위한 두 가지 엄격한 규칙으로 올바른 것은?<br>
    ① 컬럼 개수 일치, 컬럼 순서 일치<br>
    <u>② 컬럼 개수 일치, 데이터 타입 일치</u><br>
    ③ 테이블 이름 일치, 데이터 타입 일치<br>
    ④ 컬럼 이름 일치, 컬럼 개수 일치

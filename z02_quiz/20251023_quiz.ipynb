오라클이 번호 발급을 관리하여 이런 동시성 문제없이 고유 번호를 안전하게 보장함.

**Q22. (Oracle) 집합 연산자 `UNION`과 `UNION ALL`의 공통점과 가장 큰 차이점을 각각 설명하고, 어떤 연산자가 일반적으로 더 빠른지(빠른 이유) 서술하시오.**
union은 두개의 결과값을 가지고 공통을 제거하고 가져오는 과정을 거치지만,
union all은 그대로 두개의 결과값을 순차적으로 가져오고, 공통을 제거하는 과정을 거치지 않아 속도가 비교적 빠르다.
